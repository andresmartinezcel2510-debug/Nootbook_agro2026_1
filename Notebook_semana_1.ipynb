```python
# ENTRAR entra el ancho y alto de la materia
ancho = input("ingresa el ancho de tu venta en metros")
ancho = float(ancho)
alto = input("ingresa el alto de tu ventana en metros")
alto = float(alto)

#PROCESO calcular longitud de l a madera (lm) y el area del vidrio
lm= 2 * (alto * ancho) * 2.36

ar= 2 * (alto * ancho)

#SALIDA sale la longitud de la madera en (lm)y el area del vidrio en metros cuadrados
print("La longitud de la madera es", lm, "sale la longitud")
print("El area del vidrio es", ar, "metros cuadrados")
```
