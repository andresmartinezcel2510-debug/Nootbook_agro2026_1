## Ejemplos prácticos de Python

Aquí tienes tres escenarios comunes en tu campo:

### 1. Cálculo de eficiencia energética en un secador

Este pequeño código ayuda a determinar cuánta energía se requiere para eliminar el agua de un producto agrícola basándose en el calor latente.

```
# Cálculo de energía necesaria (Q = m * L)
masa_agua_a_eliminar = 50  # kg
calor_latente_vaporizacion = 2260  # kJ/kg

energia_requerida = masa_agua_a_eliminar * calor_latente_vaporizacion

print(f"La energía necesaria para el secado es: {energia_requerida} kJ")
```

### 2. Monitoreo de calidad (Control estadístico)

Supongamos que mides el pH de un lote de jugo. Este código calcula rápidamente si el promedio está dentro de los límites de control.

```
ph_muestras = [4.2, 4.3, 4.1, 4.5, 4.2]

promedio_ph = sum(ph_muestras) / len(ph_muestras)

if 4.0 <= promedio_ph <= 4.4:
    print(f"Promedio {promedio_ph}: Lote aprobado.")
else:
    print(f"Promedio {promedio_ph}: Lote fuera de especificación.")
```

### 3. Conversión de unidades (Unidad de medida crítica)

En la planta, a veces recibes datos en grados Fahrenheit y necesitas Celsius para el control de procesos.

```
def fahrenheit_a_celsius(f):
    return (f - 32) * 5/9

temp_horno_f = 212
temp_c = fahrenheit_a_celsius(temp_horno_f)

print(f"La temperatura del horno es {temp_c:.2f}°C")
```
